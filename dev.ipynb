```python
import pandas as pd
```

```python
df = pd.read_parquet("latents.parquet")
df
```

```python
reps = dict()
for label, group in df.groupby("label"):
    group = group.mean(axis=0)
    reps[label] = group
```

```python
corr = pd.DataFrame(columns=reps.keys(), index=reps.keys())
for row in reps.keys():
    for col in reps.keys():
        if row == col:
            rep = reps[row]
            points = df[df["label"] == row]
            d = points - rep
            d = d.abs().mean(axis=1)
            corr.at[row, col] = d.mean()
            continue
        dist = reps[row] - reps[col]
        dist = dist.abs().mean()
        corr.at[row, col] = dist
corr
```
